```python
import pandas as pd
import altair as alt

data = pd.read_csv("../data/ds4200_spotify_data.csv")
data.head()
```

```python
dropdown = alt.binding_select(
    options=[ 'duration_ms', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'popularity'],
    name='Y-axis attribute: '
)
ycol_param = alt.param(
    value='duration_ms',
    bind=dropdown,
    name='ycol'
)

barchart = alt.Chart(data).mark_bar().encode(
    y=alt.Y('mean(x):Q', title='Mean of Y-Axis Attribute'),
    x=alt.X('key:N', title='Key'),
    color='key:N'
).transform_calculate(
    x=f'datum[{ycol_param.name}]'
).add_params(
    ycol_param
).properties(
    title='Average Musical Attributes by Key'
)


barchart.save('barchart.html')
```

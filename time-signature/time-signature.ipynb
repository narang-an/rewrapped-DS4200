```python
import pandas as pd
import altair as alt

data = pd.read_csv("../data/ds4200_spotify_data.csv")
data.head()
```

```python
input_dropdown = alt.binding_select(options=[3, 4, 5], name='time_signature')

selection = alt.selection_point(fields=['time_signature'], bind=input_dropdown)

color = (
    alt.when(selection)
    .then(alt.Color("time_signature:N"))
    .otherwise(alt.value("lightgray"))
)

time_signature = alt.Chart(data).mark_point().encode(
    y='tempo:Q',
    x='danceability:Q',
    color=color,
).add_params(
    selection
)

time_signature.save('time-signature.html')
```
